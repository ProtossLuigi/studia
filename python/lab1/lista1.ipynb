## Zadanie 1 _(5pt)_

Napisz funkcję, który wyświetla trójkąt Pascala o zadanej wysokości
```
                1
              1   1
            1   2   1
          1   3   3   1
        1   4   6   4   1
      .....................
```

````python
def pascal_triangle(n):
    width = 1 + (n - 1) * 2
    numbers = []
    for row in range(0,n):
        for col in range(0,len(numbers)-1):
            numbers[col] += numbers[col+1]
        numbers = [1] + numbers
        for i in range(0,int((width - row * 2 - 1) / 2)):
            print(" ", end="")
        for num in numbers:
            print(num,end=" ")
        print("\n")
    return
pascal_triangle(5)
````

## Zadanie 2 _(5pt)_

Napisz funkcję `primes(n)` zwracająca listę liczb pierwszych nie większych niż $n$ np.

```
In [ ]:   primes(6)
Out [ ]: [2, 3, 5]
```

````python
def primes(n):
    prims = []
    for num in range(2,n+1):
        is_prime = True
        for prime in prims:
            if num % prime == 0:
                is_prime = False
                break
        if is_prime:
            prims.append(num)
    return prims
primes(6)
````

## Zadanie 3 _(5pt)_

Napisz funkcje, która usuwa wszystkie powtarzające się elementy z listy (tak, że każdy element występuje tylko raz) np. z listy `[1,1,2,2,2,3,3,5,5,5,4,4,4,0]` dostajemy `[1,2,3,5,4,0]`.

````python
def remove_duplicates(l):
    single = []
    for obj in l:
        if single.count(obj) == 0:
            single.append(obj)
    return single
remove_duplicates([1,1,2,2,2,3,3,5,5,5,4,4,4,0])
````

## Zadanie 4 _(10pt)_

Napisz funkcję `prime_factors(n)`, która rozkłada $n$ na czynniki pierwsze i jako wynik wraca listę par $[(p_1,\alpha_1), (p_2,\alpha_2), \ldots, (p_k, \alpha_k)]$ taką, że $p_1^{\alpha_1}p_2^{\alpha_2}\ldots p_k^{\alpha_k}$.


````python
def prime_factors(n):
    decomp = {}
    div = 2
    while n > 1:
        if n % div == 0:
            decomp[div] = decomp.get(div,0) + 1
            n /= div
        else:
            div += 1
    return list(decomp.items())
prime_factors(96)
````

## Zadanie 5 _(10pt)_

Napisz funkcję `fraczero(n)` liczącą silnie $n!$ dla $n$ od $0$ do $10000$ oraz jako wynik wraca liczbę zer na końcu $n!$.

````python
def fraczero(n):
    if n < 0 or n > 10000:
        return
    zeros = 0
    twos = 0
    fives = 0
    for i in range(2,n+1):
        num = i
        while num % 10 == 0:
            zeros += 1
            num /= 10
        while num % 5 == 0:
            fives += 1
            num /= 5
        while num % 2 == 0:
            twos += 1
            num /= 2
    if fives < twos:
        zeros += fives
    else:
        zeros += twos
    return zeros
fraczero(10000)
````

Napisz program, który generuje liczbę $20$ liczb losowych (rozkład jednostajny) od $1$ do $100$

* Wyświetl drugą największą i drugą najmniejszą wartość na liście

````python
import random
l = []
mi = 100
ma = 1
sec_min = None
sec_max = None
even = 0
summ = 0
random.seed()
for i in range(0,20):
    num = random.randint(1,100)
    l.append(num)
    summ += num
    if num > ma:
        ma = num
    elif num < ma and (sec_max is None or num > sec_max):
        sec_max = num
    if num < mi:
        mi = num
    elif num > mi and (sec_min is None or num < sec_min):
        sec_min = num
    if num % 2 == 0:
        even += 1
summ /= 20
print("liczby: ",l)
print("średnia: ",summ)
print("największa: ",ma)
print("druga największa: ",sec_max)
print("najmniejsza: ",mi)
print("druga najmniejsza: ",sec_min)
print("parzyste: ",even)
````

## Zadanie 7 _(10pt)_

Załóżmy, że mamy następującą listę: `L = ['aababacaa', 'cabaabcca', 'aaabbcbacb', 'acababbaab', ...]` wygenerowaną losowo tzn. lista zawiera losowe ciągi znaków o zadanej długości z liter `'a'`, `'b'` i `'c'`. Takie wzorce występują w wielu miejscach np. sekwencje DNA. Użytkownik zawiera swój (wzorzec) ciąg znaków z literami i pustymi miejscami oznaczonymi przez `'*'` np. `"a**a******"`. Wykorzystaj słownik z kluczami zawierającymi indeksy znaków różnych od `'*'` i wartościami liter z wzorca. Napisz program, który znajduje wszystkie sekwencje z listy, które pasuje do podanego wzorca.

````python
import random
random.seed()
L = []
for i in range(0,10):
    seq = ""
    for j in range(0,9):
        seq += random.choice("abc")
    L.append(seq)
    
def match(s):
    if len(s) != 9:
        return []
    d = {}
    for i in range(0,len(s)):
        if "abc*".count(s[i]) == 0:
            return []
        if "abc".count(s[i]):
            d[i] = s[i]
    found = []
    for seq in L:
        match = True
        for i in range(0,len(seq)):
            if seq[i] != d.get(i,seq[i]):
                match = False
                break
        if match:
            found.append(seq)
    return found
    
print(L)
print("input sequence:")
search = input()
match(search)
````
